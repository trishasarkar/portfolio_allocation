```python
len(nifty)
```

```python
final_dic={}
for i in nifty:
  df=yf.download(i, 
                 start='2010-01-01', 
                 end='2022-03-20', 
                 progress=False,
  )
  df=df.reset_index()
  final_dic[i]=list(df['Open'])
  #print(i," : ",len(list(df['Open'])))
```

```python
print(final_dic)
```

```python
df_final=pd.DataFrame.from_dict(final_dic)
df_final
```

```python
# Calculate the annual mean returns and variances
daily_returns = df_final.pct_change()
annual_mean_returns = daily_returns.mean()*252
annual_return_variance = daily_returns.var() * 252
```

```python
# Create a new dataframe
df2=pd.DataFrame(df_final.columns, columns=['Stock_Symbols'])
df2['Variances']= annual_return_variance.values
df2['Returns']= annual_mean_returns.values
df2
```

```python
# Use the Elbow method to determine the number of clusters to use the group of stocks
# Get and store the annual returns and annual variances
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X= df2[['Returns','Variances']].values
inertia_list=[]
for k in range(2,45):
  # Create and train the model
  kmeans= KMeans(n_clusters=k)
  kmeans.fit(X)
  inertia_list.append(kmeans.inertia_)

#Plot the data
plt_1=plt.figure(figsize=(15,5))
plt.plot(range(2,45),inertia_list)
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia or Sum Squared Error ( SSE )')
plt.xticks(np.arange(2,45))
plt.show()
```

```python
# Get and show the labels / groups

kmeans= KMeans(n_clusters=4,random_state=0).fit(X)
labels= kmeans.labels_
labels
```

```python
df2['Cluster Labels']=labels
df2
```

```python
# Plot and show the difference clusters
plt.scatter(X[:,0],X[:,1],c=labels,cmap='rainbow')
plt.title('K-Means Plot')
plt.xlabel('Returns')
plt.ylabel('Variances')
plt.colorbar()
plt.show()
```

```python
name_med=[]
for i in df2.loc[df2['Cluster Labels'] == 3]['Stock_Symbols']:
  name_med.append(i)
name_med
```

```python
# Return Per unit of Risk
Avg_daily / (daily_simple_return.std() * np.sqrt(252)) *100
```

The higher this ratio, the better it is. Hence, Bajaj Finance has the best Return to Risk ratio, Abbott India Ltd having the lowest. After adjusting for a risk-free rate, this ratio is also called Sharpe Ratio, a measure of risk-adjusted return. It describes how much excess return you receive for the volatility of holding a riskier asset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from datetime import date

from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import  risk_models
from pypfopt import expected_returns
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
```

Covariance is a measurement of the spread between numbers in a dataset, i.e., it measures how far each number in the data set is from the mean. The higher the variance of an asset price, the higher the risk asset bears and a higher return. The most commonly-used risk model is the covariance matrix, which describes asset volatilities and their co-dependence. This is important because one of the principles of diversification is that risk can be reduced by making many uncorrelated bets (Correlation is nothing but the standardized version of covariance)

```python
mean = expected_returns.mean_historical_return(df_disp)
S = risk_models.sample_cov(df_disp)# for sample covariance matrix
S
```

```python
plt.style.use('ggplot')
fig = plt.figure()
sb.heatmap(S,xticklabels=S.columns, yticklabels=S.columns,
cmap='RdBu_r', annot=True, linewidth=0.5)
print('Covariance between daily simple returns of stocks in your portfolio')
plt.show(fig)
```

Sharpe ratio describes that how much excess return you receive for the extra volatility you endure for holding a risky asset. After determining the Mean Historical Returns, we're assessing the efficient frontier by setting the Optimization condition as "Max Sharpe" to have weights in such a way that will maximize the Sharpe Ratio. Then, we're cleaning the raw weights, rounding them off & setting any weights whose absolute values are below the cutoff to 0%

```python
ef = EfficientFrontier(mean,S)
weights = ef.max_sharpe() #for maximizing the Sharpe ratio #Optimization
cleaned_weights = ef.clean_weights() #to clean the raw weights
# Get the Keys and store them in a list
labels = list(cleaned_weights.keys())
# Get the Values and store them in a list
values = list(cleaned_weights.values())
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.0f%%')
print('Portfolio Allocation')
plt.show(fig)
```

```python
ef.portfolio_performance(verbose=True)
```
